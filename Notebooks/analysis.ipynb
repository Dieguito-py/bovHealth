# **Dados Bruto**

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

file_path = '../data/raw/dados2.csv'

if not os.path.exists(file_path):
    print("File not found:", file_path)
else:
    df = pd.read_csv(file_path)

plt.subplot(1, 3, 1)
plt.boxplot(df['x'])
plt.ylim(-1, 1)
plt.title('Boxplot de X')

plt.subplot(1, 3, 2)
plt.boxplot(df['y'])
plt.ylim(-1, 1)
plt.title('Boxplot de Y')

plt.subplot(1, 3, 3)
plt.boxplot(df['z'])
plt.ylim(-1, 1)
plt.title('Boxplot de Z')

plt.suptitle('Boxplots das Leituras do Acelerômetro')
plt.show()

atividades = df['atividade'].unique()

plt.figure(figsize=(15, 10))

for i, atividade in enumerate(atividades, 1):
    plt.subplot(len(atividades), 3, i * 3 - 2)
    plt.boxplot(df[df['atividade'] == atividade]['x'])
    plt.ylim(-1, 1)
    plt.title(f'X - {atividade}')
    
    plt.subplot(len(atividades), 3, i * 3 - 1)
    plt.boxplot(df[df['atividade'] == atividade]['y'])
    plt.ylim(-1, 1)
    plt.title(f'Y - {atividade}')
    
    plt.subplot(len(atividades), 3, i * 3)
    plt.boxplot(df[df['atividade'] == atividade]['z'])
    plt.ylim(-1, 1)
    plt.title(f'Z - {atividade}')

plt.suptitle('Boxplots das Leituras do Acelerômetro por Atividade')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# Visualização Dados Brutos

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '../data/raw/dados3.csv'
data = pd.read_csv(file_path)

print(data.head())

plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(data['x'], color='r')
plt.title('Dados do Acelerômetro - Eixo X')
plt.ylim([-2, 2])
plt.ylabel('Aceleração (g)')

plt.subplot(3, 1, 2)
plt.plot(data['y'], color='g')
plt.title('Dados do Acelerômetro - Eixo Y')
plt.ylim([-2, 2])
plt.ylabel('Aceleração (g)')

plt.subplot(3, 1, 3)
plt.plot(data['z'], color='b')
plt.title('Dados do Acelerômetro - Eixo Z')
plt.ylim([-2, 2])
plt.xlabel('Amostras')
plt.ylabel('Aceleração (g)')

plt.tight_layout()
plt.show()

activities = data['atividade'].unique()
print(f'Atividades identificadas: {activities}')

plt.figure(figsize=(12, 8))

for activity in activities:
    subset = data[data['atividade'] == activity]
    plt.plot(subset['x'], label=f'Atividade {activity} - X')
    plt.plot(subset['y'], label=f'Atividade {activity} - Y')
    plt.plot(subset['z'], label=f'Atividade {activity} - Z')

plt.title('Dados do Acelerômetro por Atividade')
plt.ylim([-2, 2])
plt.xlabel('Amostras')
plt.ylabel('Aceleração (g)')
plt.legend()
plt.show()
```
